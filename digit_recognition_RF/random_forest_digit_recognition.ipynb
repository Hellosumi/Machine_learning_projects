Step 1:Import required libraries

```python
# Importing necessary libraries
import pandas as pd
```

Step 2:Choose Dataset from local Directory

Step 3: Load dataset from the given dataset file

```python
# Loading the dataset
dataset = pd.read_csv('trainset.csv')
```

Step 4: Summarise dataset
We can see the dataset which we have uploaded

```python
# Display the shape of the dataset and the first row
print(dataset.shape)
print(dataset.head(5))
```

Step 5: Segregate data into X(input) & Y(output)

```python
# Splitting the dataset into features (X) 
x=dataset.iloc[:,1:] ##the dataset loaded from reverse direction
print(x)
print(x.shape)
```

```python
# Splitting the dataset into target variable (y)
y=dataset.iloc[:,0]
print(y)
print(y.shape)
```

step 6: splitting dataset into test and train... 80% data for train and 20% data for test.

```python
# Splitting the data into training and testing sets
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split (x,y, test_size=0.20, random_state=0)
```

Step 7:Training the model

```python
# Importing and fitting the RandomForestClassifier
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
model.fit(x_train, y_train)
```

```python
# Predicting the test set results
y_pred=model.predict(x_test) ##store the output value of model prediction in "y_pred" variable
```

Step 8: Calculate the model's accuracy using accuracy_score.

```python
# Calculating and printing the accuracy of the model
from sklearn.metrics import accuracy_score
print("Accuracy of the model is :{0}%".format(accuracy_score(y_test,y_pred)*100))
```

Step 9: Model prediction

```python
# Displaying predictions with their actual values
index = 1200
print(f'Actual output : {y_test.iloc[index]}') # Use .iloc for positional indexing
print(f'Predicted output : {y_pred[index]}')
```
